The similarity is measured as [Euclidean distance](https://en.wikipedia.org/wiki/Euclidean_distance) between vectors of players' characteristics (described in the task above). Put their names in a vector in ascending order by Euclidean distance and sorted by `player_name` if the distance is the same<br>

```python
# player_att_data.loc[:, ~player_att_data.columns.isin(['id', 'player_fifa_api_id', 'player_api_id', 'date', 'preferred_foot', \
#                                                           'attacking_work_rate', 'defensive_work_rate'])]
neymar_api_id = player_data[player_data.player_name == 'Neymar'].iloc[0]['player_api_id']
player_att_data_10 = player_att_data[:].dropna()
player_att_data_10['date'] = pd.to_datetime(player_att_data_10.date)
neymar_attr = player_att_data_10[player_att_data_10.player_api_id == neymar_api_id].sort_values(by='date', ascending=False).head(1) \
                                                                .loc[:, ~player_att_data_10.columns.isin(['id', 'player_fifa_api_id', 'player_api_id', 'date', 'preferred_foot', \
                                                          'attacking_work_rate', 'defensive_work_rate'])]
last_measures_dates = player_att_data_10.groupby('player_api_id')['date'].max()
latest_measures = pd.merge(last_measures_dates, player_att_data_10, on=['player_api_id', 'date']).loc[:, ~player_att_data.columns.isin(['player_fifa_api_id', 'date', 'preferred_foot', \
                                                          'attacking_work_rate', 'defensive_work_rate'])]
latest_measures['euclidian_distance'] = latest_measures.loc[:, 'overall_rating':'gk_reflexes'].sub(neymar_attr.squeeze()).pow(2).sum(axis=1)
closest_players_ids = latest_measures.sort_values(by='euclidian_distance').head(6).iloc[1:]['player_api_id']
neymar_similarities = pd.merge(closest_players_ids, player_data, on='player_api_id').iloc[:]['player_name'].tolist()
```

**Task 12 (1 point).** Find a team having the most matches (both home and away!) in the **Germany 1. Bundesliga** in **2008/2009** season. Return number of matches.

```python
away_matches = match_data[(match_data.season == '2008/2009') & (match_data.league_id == bundesleague_id)] \
                        .groupby('away_team_api_id').size()
home_matches = match_data[(match_data.season == '2008/2009') & (match_data.league_id == bundesleague_id)] \
                        .groupby('home_team_api_id').size()
team_most_matches_bundesliga_2008_2009 = away_matches.add(home_matches).sort_values(ascending=False).iloc[0]
```

**Task 13 (1 point).** Count total number of **Arsenal** matches (both home and away!) in the **2015/2016** season which they have won. <br><br>
**Note:** Winning a game means scoring **more** goals than an opponent.

```python
arsenal_id = team_data[team_data.team_long_name == 'Arsenal'].iloc[0]['team_api_id']
num_home_won_mathes = match_data[(match_data.season == '2015/2016') & (match_data.home_team_api_id == arsenal_id) & \
                                 (match_data.home_team_goal > match_data.away_team_goal)].shape[0]
num_away_won_mathes = match_data[(match_data.season == '2015/2016') & (match_data.away_team_api_id == arsenal_id) & \
                                 (match_data.away_team_goal > match_data.home_team_goal)].shape[0]
arsenal_won_matches_2015_2016 = num_home_won_mathes + num_away_won_mathes
```

**Task 14 (2 points).** Find a team with the highest win rate in the **2015/2016** season. Win rate means won matches / all matches. If there are several teams with the highest win rate return the first by name in lexical order

```python
won_at_home = match_data[(match_data.season == '2015/2016') & (match_data.home_team_goal > match_data.away_team_goal)] \
                        .groupby('home_team_api_id').size()
won_away = match_data[(match_data.season == '2015/2016') & (match_data.home_team_goal < match_data.away_team_goal)] \
                        .groupby('away_team_api_id').size()
all_at_home = match_data[match_data.season == '2015/2016'].groupby('home_team_api_id').size()
all_away = match_data[match_data.season == '2015/2016'].groupby('away_team_api_id').size()
id_team_highest_winrate = won_at_home.add(won_away).divide(all_at_home.add(all_away)).sort_values(ascending=False).index[0]
team_highest_winrate_2015_2016 = team_data[team_data.team_api_id == id_team_highest_winrate].iloc[0]['team_long_name']
```

**Task 15 (2 points).** Determine the team with the maximum days' gap between matches in **England Premier League 2010/2011 season**. Return number of days in that gap. <br>
**Note**: a *gap* means the number of days between two consecutive matches of the same team.

```python
eng_league_id = league_data[league_data.name == 'England Premier League'].iloc[0]['id']
matches_eng_league = match_data[(match_data.season == '2010/2011') & (match_data.league_id == eng_league_id)]
home_matches_eng = matches_eng_league[['date','home_team_api_id']].rename(columns={'home_team_api_id': 'team_api_id'}) 
away_matches_eng = matches_eng_league[['date','away_team_api_id']].rename(columns={'away_team_api_id': 'team_api_id'})
team_date = pd.concat([home_matches_eng, away_matches_eng])
team_date['date'] = pd.to_datetime(team_date['date'])
gaps = team_date.sort_values(by=['team_api_id', 'date'], ascending=(True, True)).diff()
highest_gap_england_2010_2011 = gaps[gaps.team_api_id == 0].sort_values(by='date', ascending=False).iloc[0]['date'].days
```
